```python
import numpy as np
import copy
import matplotlib.pyplot as plt
import cv2
fx=4000.
fy=4000.
cx=554
cy=303
K = np.array([[fx,0,cx],
        [0,fy,cy],
        [0.,0.,1]])
k1=-10.
k2=0.
k3=0.
k4=0.
D = np.array([k1,k2,k3,k4])
Knew=copy.copy(K)
img = cv2.imread("fisheye_original.jpg")

point = np.array([[[300.,300.],[300,400],[400,400],[400,300]]])
img_out = cv2.fisheye.undistortImage(img,K=K, D=D,Knew=Knew)

# save the img
cv2.imwrite("salida"+'.jpg',img_out,[100])
plt.imshow(img_out)
```
